# Exercise 1

```python
variance = np.zeros(90)
for n in range(10,100):
    X = np.random.randint(1,7,(1000,n))
    Xbar = np.mean(X,axis=1)
    variance[n-10] = np.var(Xbar)
plt.plot(np.arange(10,100),variance,'bo');
```

```python
var = np.zeros(9)
for sig in range(1,10):
    X = np.random.normal(0,sig,(1000,10))
    Xbar = np.mean(X,axis=1)
    var[sig-1] = np.var(Xbar)
plt.plot(np.arange(1,10),var,'bo')
```

# CLT

```python

def clt(n):
    X = np.random.rand(10000,n)
    Xsum = np.sum(X,axis=1)
    print(Xsum,np.mean(Xsum),np.var(Xsum))
    plt.hist(Xsum)
```

```python
clt(1)
```

```python
clt(10)
```

```python
clt(100)
```

```python
clt(1000)
```

```python
clt(10000)
```
