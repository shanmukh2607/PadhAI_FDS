```python
url = 'https://api.covid19india.org/states_daily.json'
```

```python
import urllib.request
```

```python
urllib.request.urlretrieve(url, 'data.json')
```

```python
import numpy as np
import pandas as pd
```

```python
covid_data = pd.read_json('data.json')
```

```python
covid_data
```

```python
import json
with open('data.json') as f:
    data = json.load(f)
data
```

```python
data = data['states_daily']
```

```python
data
```

```python
covid_data = pd.json_normalize(data)
```

```python
covid_data
```

```python
df = covid_data
df.date = pd.to_datetime(df.date)
```

```python
df = df[df['status']== 'Confirmed']
df
```

```python
df.drop('status',axis=1,inplace = True)
```

```python
df.set_index('date',inplace = True)
```

```python
df.drop('dateymd',axis =1,inplace = True)
df=df.apply(pd.to_numeric)
```

```python
df
```

```python
df.tg
```

# Styling

```python
df = df.tail(7)
```

```python
df.style
```

```python
def color_negative_red(x):
    color = 'red' if x<0 else 'white'
    return 'color:' + color
```

```python
df.drop('mh',axis=1,inplace = True)
df.style.applymap(color_negative_red)
```

```python
df.style.highlight_max(color='red')
```

```python
df.drop(['dn','un'],axis=1,inplace=True)
```

```python
df.style.highlight_max(color='red').highlight_min(color='green')
```

```python
def bold_maxval(x):
    gg = (x==x.max())
    return ['font-weight:bold' if y else '' for y in gg]
```

```python
df.drop('tt',axis=1,inplace=True)
df.style.apply(bold_maxval).highlight_max(color ='red',axis=1)
```

```python
df.loc[:,'ap':'tn'].style.background_gradient(cmap='Reds')
```
